```python
def and_gate(
    x1,
    x2,
    w1: float = 0.5,
    w2: float = 0.5,
    theta: float = 0.7,
  ):
  y = x1 * w1 + x2 * w2
  if y <= theta:
    return 0
  else:
    return 1
```

```python
and_gate(0, 0)
```

```python
import numpy as np

def and_gate(
    x1,
    x2,
    w1: float = 0.5,
    w2: float = 0.5,
    theta: float = 0.7,
) -> int:
  x = np.array([x1, x2])
  w = np.array([w1, w2])
  b = theta * -1
  y = np.sum(x * w)  + b
  if y <= 0: return 0
  else: return 1
```

```python
and_gate(0, 0)
```

```python
import numpy as np

def nand_gate(
    x1,
    x2,
    w1: float = 0.5,
    w2: float = 0.5,
    theta: float = 0.7,
) -> int:
  x = np.array([x1, x2])
  w = np.array([-w1, -w2])
  b = theta
  y = np.sum(x * w)  + b
  if y <= 0: return 0
  else: return 1
```

```python
nand_gate(0, 0)
```

```python
import numpy as np

def or_gate(
    x1,
    x2,
    w1: float = 0.5,
    w2: float = 0.5,
    theta: float = 0.7,
) -> int:
  x = np.array([x1, x2])
  w = np.array([-w1, -w2])
  b = theta * -1
  y = np.sum(x * w)  + b
  if y <= 0: return 0
  else: return 1
```

```python
import numpy as np

def xor_gate(
    x1,
    x2,
    w1: float = 0.5,
    w2: float = 0.5,
    theta: float = 0.7,
) -> int:
  s1 = nand_gate(x1, x2, w1, w2, theta)
  s2 = or_gate(x1, x2, w1, w2, theta)
  y = and_gate(s1, s2, w1, w2, theta)
  return y
```

```python
print(or_gate(0, 0), xor_gate(0, 0))
```
